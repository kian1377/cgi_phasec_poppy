# Import the necessary packages and define the path to the phaseb_data directory as well as the new directory you would like to save the altered files to. 

```python
from astropy.io import fits
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'image.origin': 'lower'})
import os
from pathlib import Path

import roman_phasec_proper as phasec
spcdir_old = Path(phasec.data_dir+'/spc_20200617_spec')

import cgi_phasec_poppy
spcdir = cgi_phasec_poppy.data_dir/'spc-spec' # path to new directory

from cgi_phasec_poppy import misc

pupil_diam = 2.363114*u.m
pupil_diam_pix = 1000*u.pixel
```

# Add pupil scales to the pupil files.

```python
pupil_fnames = ['pupil_SPC-20200617_1000.fits']

for fname in pupil_fnames:
    spc730pup = fits.open(spcdir_old/fname)
    spc730pup_data = spc730pup[0].data
    spc730pup.close()
    
    arr = np.zeros(shape=(1002,1002))
    arr[1:,1:] = spc730pup_data
    spc730pup_data = arr
    
    hdr = fits.Header()
    hdr['PUPLSCAL'] = pupil_diam.value/pupil_diam_pix.value
    hdr['PIXELSCL'] = pupil_diam.value/pupil_diam_pix.value
    spc730pup_new = fits.PrimaryHDU(data=spc730pup_data, header=hdr)
    spc730pup_new.writeto(spcdir/fname, overwrite=True)
    print(spcdir/fname)
    print(pupil_diam/pupil_diam_pix, spc730pup_data.shape)
    misc.myimshow(spc730pup_data, pxscl=pupil_diam/pupil_diam_pix)
```

# Add pupil scales to the pupil mask files.

```python
spm_fnames = ['SPM_SPC-20200617_1000_rounded9.fits','SPM_SPC-20200617_1000_rounded9_rotated.fits']
spm_sampling = 1.699867952187158e-05*u.m/u.pixel

for fname in spm_fnames:
    spc730spm = fits.open(spcdir_old/fname)
    spc730spm_data = spc730spm[0].data
    spc730spm.close()
    
    arr = np.zeros(shape=(1002,1002))
    arr[1:,1:] = spc730spm_data
    spc730spm_data = arr
    
    hdr = fits.Header()
    hdr['PIXELSCL'] = spm_sampling.value
    spc730spm_new = fits.PrimaryHDU(data=spc730spm_data, header=hdr)
    spc730spm_new.writeto(spcdir/fname, overwrite=True)
    print(spcdir/fname)
    print(spc730spm_data.shape)
    misc.myimshow(spc730spm_data, pxscl=spm_sampling)
```

# Add pixel scales to the FPM files.

```python
fpm_fnames = ['fpm_0.05lamD.fits']

for fname in fpm_fnames:
    spc730fpm = fits.open(spcdir_old/fname)
    spc730fpm_data = spc730fpm[0].data
    spc730fpm.close()
    fpm_pxscl_lamD = 0.05
    fpm_lambda_m = 730e-9
    fpm_pxscl_as = (fpm_pxscl_lamD*(fpm_lambda_m/pupil_diam.value)*u.radian/u.pix).to(u.arcsec/u.pix)
    print(fpm_lambda_m, fpm_pxscl_lamD, fpm_pxscl_as)
    
#     arr = np.zeros(shape=(382,382))
#     arr[1:,1:] = spc730fpm_data
#     spc730fpm_data = arr
    
    print(spcdir/fname)
    print(spc730fpm_data.shape)
    hdr = fits.Header()
    hdr['PIXELSCL'] = fpm_pxscl_as.value
    hdr['PIXSCLLD'] = fpm_pxscl_lamD
    hdr['WAVELENC'] = fpm_lambda_m
    hdr['PUPLDIAM'] = pupil_diam.value
    spc730fpm_new = fits.PrimaryHDU(data=spc730fpm_data, header=hdr)
    spc730fpm_new.writeto(str(spcdir/fname), overwrite=True)
    
    misc.myimshow(spc730fpm_data, pxscl=fpm_pxscl_as)
    
    
    
    
```

# Add pupil scales to the lyot stop files.

```python
ls_fnames = ['LS_SPC-20200617_500.fits','LS_SPC-20200617_1000.fits']
ls_samplings = [1.699867952187158e-05*2*u.m/u.pixel, 1.699867952187158e-05*u.m/u.pixel] 

for i,fname in enumerate(ls_fnames):
    spc730ls = fits.open(spcdir_old/fname)
    spc730ls_data = spc730ls[0].data
    spc730ls.close()
    
    if i==0:
        arr = np.zeros(shape=(502,502))
        arr[1:,1:] = spc730ls_data
        spc730ls_data = arr
    else: 
        arr = np.zeros(shape=(1002,1002))
        arr[1:,1:] = spc730ls_data
        spc730ls_data = arr
        
    
    hdr = fits.Header()
    hdr['PIXELSCL'] = ls_samplings[i].value
    spc730ls_new = fits.PrimaryHDU(data=spc730ls_data, header=hdr)
    spc730ls_new.writeto(spcdir/fname, overwrite=True)
    print(spcdir/fname)
    print(spc730ls_data.shape)
    misc.myimshow(spc730ls_data, pxscl=ls_samplings[i])
```
