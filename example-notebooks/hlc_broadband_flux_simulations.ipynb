# Calculate flux of the star using its blackbody spectrum and solid angle.


## Assuming 47 UMa

```python
T = 5887*u.K
d = (14.06*u.parsec).to(u.m)
R_uma47 = 1.172*R_sun

SA_uma47 = np.pi * (R_uma47/d)**2 * u.sr
display(SA_uma47)

spec_uma47 = 2*np.pi*h*c**2/(lambdas**5)/(np.exp(h*c/(k_B*T*lambdas)) - 1)/u.sr
spec_uma47 = spec_uma47.to(u.J/u.s/u.sr/u.m**2/u.nm) * SA_uma47 # multiply by solid angle to get irradiance
display(spec_uma47.unit)

fig,ax = plt.subplots(1,1,dpi=125)
ax.plot(lambdas, spec_uma47)
ax.set_xlabel('$\lambda$ [nm]')
ax.set_ylabel('$W/m^2/sr/nm$')
```

## Assuming $\zeta$ Pup

```python
T = 40000*u.K
d = (330*u.parsec).to(u.m)
R_zpup = 20*R_sun

SA_zpup = np.pi * (R_zpup/d)**2 * u.sr
display(SA_zpup)

spec_zpup = 2*np.pi*h*c**2/(lambdas**5)/(np.exp(h*c/(k_B*T*lambdas)) - 1)/u.sr
spec_zpup = spec_zpup.to(u.J/u.s/u.sr/u.m**2/u.nm) * SA_zpup # multiply by solid angle to get irradiance
display(spec_zpup)

fig,ax = plt.subplots(1,1,dpi=125)
ax.plot(lambdas, spec_zpup)
ax.set_xlabel('$\lambda$ [nm]')
ax.set_ylabel('$W/m^2/sr/nm$')
```

# Use numerical integration to get the approximate flux per propagation wavelength

```python
# choose which spectrum to use
spec = spec_uma47
# spec = spec_zpup

nlam = 3
bandwidth = 0.10
minlam = wavelength_c * (1 - bandwidth/2)
maxlam = wavelength_c * (1 + bandwidth/2)
wavelengths = np.linspace( minlam, maxlam, nlam )

del_waves = wavelengths[1]-wavelengths[0]
fluxes = []
for i in range(nlam):
    wave = wavelengths[i].to(u.nm)
    min_wave = wave-del_waves/2
    max_wave = wave+del_waves/2
    inds = np.where((lambdas>min_wave)&(lambdas<max_wave))
    
    lambdas_domain = lambdas[inds].value
    spec_range = spec[inds].value
    
    fluxes.append(simpson(spec_range, lambdas_domain)*u.J/u.s/u.m**2) # integrate over the wave band
    
    E = h*c/wave.to(u.m) / u.photon # energy of the photon for this wavelength
    fluxes[i] /= E # convert to photons/s/m**2
    
display(fluxes)
```
